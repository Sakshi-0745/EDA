```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([0,6])
y = np.array([0,300])
plt.plot(x,y)
plt.show()
```

```python
a = np.array([3,8,9,1,7,3])
plt.plot(a,linestyle='dotted')
plt.show()
```

```python
a = np.array([3,8,9,1,7,3])
plt.plot(a,'*:y')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
a = np.array([0,8])
b = np.array([0,350])
plt.subplot(1,2,2)
plt.plot(a,b)
plt.show()
```

```python
x = np.array([4,5,6,1,8,2,3,10,9,3])
y = np.array([5,2,0,2,7,6,0,2,6,0])
plt.scatter(x,y)
x = np.array([11,28,5,20,2,45,2,41,4,52])
y= np.array([12,5,2,10,76,2,93,85,30,52])
plt.scatter(x,y)
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([5,3,6,1])
plt.hist(y)
plt.show()
```
